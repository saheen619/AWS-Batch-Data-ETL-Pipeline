```python
import requests
import json

url = f"https://www.consumerfinance.gov/data-research/consumer-complaints/search/api/v1/" \
      f"?field=all&format=json" \
      f"&date_received_max=2023-06-08&date_received_min=2023-06-01"

headers = {'User-agent': 'consumer complaints extractor'}
response = requests.get(url, headers)
data = json.loads(response.content)
consumer_complaints_data = []

def extract_consumer_complaints_data(url):
    for item in data:
        if "_source" in item.keys() :
            consumer_complaints_data.append(item["_source"])
    return item
```
